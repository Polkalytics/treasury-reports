```python
import requests
import json
import pandas as pd
```

```python
# load proposals
url = "https://polkadot-api.dotreasury.com/proposals?page=0&page_size=500&status=Awarded"
# Step 1: Fetch the URL's content
response = requests.get(url)
response_json = response.json()

# normalize into df
df = pd.DataFrame(response_json)
df = pd.json_normalize(df['items'])
df
```

```python
def get_proposal(proposal_index):
  url = "https://polkadot.subsquare.io/api/treasury/proposals/" + proposal_index
  response = requests.get(url)
  response_json = response.json()
  return response_json
```

```python
# Connect to Google Drive
from google.colab import auth
import gspread
from google.auth import default
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
# Read the Treasury Spreadsheet
import pandas as pd
worksheet = gc.open('Treasury').sheet1
rows = worksheet.get_all_values()
df = pd.DataFrame(rows[1:], columns=rows[0])
df
```

```python
# extract year
# Remove the first character from the 'Index' column
df['Index'] = df['proposalIndex']
df["year"] = df['Propose Time'].str[0:4].astype(int)
# extract USD & DOT
df["USD"] = [float(x.split('$')[1].replace('.','').replace(',','.')) for x in df["Value"]]
df["DOT"] = [float(x.split('\n')[0].replace('.','').replace(',','.')) for x in df["Value"]]
df["link"] = ["https://polkadot.subsquare.io/treasury/proposals/" + x for x in df["Index"]]
df['Description'] = df['Description'].str.split('\n').str[0]
df["HYPERLINK"] = None
df["Category"] = None
df["Subcategory"] = None
df = df[["Index", "HYPERLINK", "Category", "Subcategory", "Description", "Propose Time", "Beneficiary", "Value", "Status", "year", "USD", "DOT", "link"]]
```

```python
file_path = "/content/sample_data.xlsx"
df.to_excel(file_path, index=False)
```

```python
df_spend = df[["year", "USD", "DOT"]].groupby(["year"]).sum()
df_spend["USD"] = df_spend["USD"].map('{:,.0f} USD'.format)
df_spend["DOT"] = df_spend["DOT"].map('{:,.0f} DOT'.format)
df_spend
```
